```python
from bs4 import BeautifulSoup as bs4
import requests
import pandas as pd
import unicodedata
from lxml import html 
import re
import numpy as np
```

```python
pages=[]
prices=[]
streets=[]
descriptions=[]
zhks=[]
head=[]
titles=[]
cena=[]
```

```python
pages_to_scrape=10
```

```python
for i in range(2,pages_to_scrape+1):
    url = ('https://krisha.kz/prodazha/kvartiry/nur-sultan/?areas=&bounds=&das[novostroiki]=1&page={}').format(i)
    pages.append(url)
```

# Цена

```python
for item in pages:
    page = requests.get(item)
    soup = bs4(page.text, 'html.parser')
    divTag = soup.findAll("div", attrs={"class": "a-card__descr"})
    for tag in divTag:
        tdTags = tag.findAll("div",attrs= {"class": "a-card__price"})
        for tag in tdTags:
            prices.append(tag.text)
```

```python
for i in prices:
    string = i.replace(u'\xa0', u'')
    x=re.findall('\d+', string )
    
    cena.append(x)
    
```

```python
len(cena)
```

```python
stoimost=[]
```

```python
for i in cena:
    for j in i:
        x=int(j)
        stoimost.append(x)
    
```

# Заголовок

```python
for item in pages:
    page = requests.get(item)
    soup = bs4(page.text, 'html.parser')
    divHead = soup.findAll("div", attrs={"class": "a-card__descr"})
    for tag in divHead:
        tdHeads = tag.findAll("a",href=True,attrs= {"class": "a-card__title"})
        for tag in tdHeads:
            head.append(tag.text)
```

```python
head
```

```python
len(head)
```

```python
area=[]
komnaty=[]
```

```python
for i in head:
    
    x=i.split()
    komnaty.append(int(x[0][:1]))
    area.append(float(x[2]))
```

# Текст

```python
for item in pages:
    page = requests.get(item)
    soup = bs4(page.text, 'html.parser')
    divStreet = soup.findAll("div", attrs={"class": "a-card__descr"})
    for tag in divStreet:
        tdStreets = tag.findAll("div",attrs= {"class": "a-card__text-preview"})
        for tag in tdStreets:
            descriptions.append(tag.text)
```

# ЖК

```python
for item in pages:
    page = requests.get(item)
    soup = bs4(page.text, 'html.parser')
    divStreet = soup.findAll("div", attrs={"class": "a-card__descr"})
    for tag in divStreet:
        tdStreets = tag.findAll("div",attrs= {"class": "a-card__complex-info"})
        for tag in tdStreets:
            zhks.append(tag.text)
```

```python
zhks
```

```python
zhk_name=[]
zastroishik=[]
```

```python
for i in zhks:
    x=i.split()
    c=' '.join(x[x.index('ЖК')+1:x.index('от')])
    zhk_name.append(c)
    y=' '.join(x[x.index('застройщика')+1:])
    zastroishik.append(y)
```

```python
data={'ЖК': zhk_name,"Застройщик":zastroishik, 'Площадь (м2)': area, 'Количество комнат':komnaty, "Информация":descriptions,'Цена (тнг)': stoimost}
```

```python
df=pd.DataFrame(data=data)
```

```python
df
```

```python
df["Высота потолков (м)"]=np.nan
df["Кухня"]=np.nan
df["Жил площадь (м2)"]=np.nan
```

```python
ceiling_height=[]
index=0
for i in df["Информация"]:
    x=i.split()
    if "площадь" in x:
        z=x[x.index('площадь')+1]
        
        df.loc[index,'Жил площадь (м2)']=float(z)
    if "кухня" in x:
        z=x[x.index('кухня')+1]
        
        df.loc[index,'Кухня']=float(z)
    if "потолки" in x:
        z=x[x.index('потолки')+1]
        y=z.replace("м.,",'')
        df.loc[index,'Высота потолков']=float(y)
    index+=1
    
        
    
```
