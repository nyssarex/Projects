```python
#box plot overallqual/saleprice
var = 'Количество комнат'
df = pd.concat([data['Цена (тнг)'], data[var]], axis=1)
f, ax = plt.subplots(figsize=(8, 6))
fig = sns.boxplot(x=var, y="Цена (тнг)", data=df)
fig.axis(ymin=4000000, ymax=100000000);
```

Перед тем как посмотреть на график сверху, давайте ознакомимся что такое box plot и что он показывает.
Нижний и верхний прочерк показывает минимальное и максимальное значение,соответственно 
Нижняя и верхняя граница прямоугольника показывает 25% квартиль и 75% квартиль, соответсвенно.
То есть Верхний квартиль это значение цены за квартиру при которой 75% квратир имееют цену меньше, а
Нижний квартиль это значение цены за квартиру при которой 25% квартир имеют значение меньше. 
Если сортировать цены за квартиры по возрастанию и выбрать центральное значение, это будет медианой.


Можно заметить точки в графике 2х комнатных квартир, которые уходят за пределы мин и макс. Эти значения называются выбросами. То есть застройщик завысил цену по сравнению с общей картиной.

```python
var = 'Высота потолков (м)'
df = pd.concat([data['Цена (тнг)'], data[var]], axis=1)
f, ax = plt.subplots(figsize=(16, 8))
fig = sns.boxplot(x=var, y="Цена (тнг)", data=df)
fig.axis(ymin=4000000, ymax=100000000);
```
